```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
import time
```

```python
df1 = pd.read_csv('5final_Dermatologist.csv')
df1
```

```python
listp=list(df1['0'])
```

```python
listp
```

```python
pedia=[]
driver=webdriver.Firefox()
for i in listp:
    url=f'https://www.practo.com{i}'
    driver.get(url)
    soup=BeautifulSoup(driver.page_source,'html.parser')
    try:

        dentist1 = soup.find_all('div',{'class':'pure-g g-card'})[0]
        name = dentist1.find_all('h1',{'class':'c-profile__title u-bold u-d-inlineblock'})[0].text
        qualification = dentist1.find_all('p',{'class':'c-profile__details'})[0].text
        years = dentist1.find('div',{'class':'c-profile__details'}).find_all('h2')[-1].text
        dentist2 = soup.find_all('div',{'class':'c-profile--sections'})[0]
        Location = dentist2.find('h4',{'class':'c-profile--clinic__location'}).text
        try:
            dp_score = dentist1.find('span',{'class':'u-green-text u-bold u-large-font'}).text
        except:
            dp_score = 0
        try:
            npv= dentist1.find('span',{'class':'u-smallest-font u-grey_3-text'}).text
        except:
            npv=0
        try:
            consultation_fee = dentist2.find('span',{'data-qa-id':'consultation_fee'}).text
        except:
            consultation_fee=0
        tabledict={'Name': name,'Qualification':qualification,'Years_Of_Experience': years,'Speciality_of_doctor':'Dermatologist','Location':Location,'dp_score':dp_score,'npv':npv,'consultation_fee':consultation_fee}
        
        pedia.append(tabledict)
    except:
        pass
```

```python
table=pd.DataFrame(pedia)
table
```

```python
table.to_csv('Dermatologist.csv')
```
